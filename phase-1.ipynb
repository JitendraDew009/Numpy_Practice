## List vs Numpy


```python
py_list = [1,2,3]
print("Pyhton list multiplication", py_list*2)
```

```python
np_array = np.array([1, 2, 3])
print("Python array multiplication", np_array*2)
```

```python
import time
start = time.time()
py_list = [i*2 for i in range(1000000)]
print("\n List operation time :", time.time() - start) # 0.10156512260437012

start = time.time()
np_array = np.arange(1000000)*2 
print("\n Numpy operation time :", time.time() - start) # 0.005458354949951172
```

## Vector, Matrix, and Tensor

```python
vector = np.array([1, 2, 3])
print("Vector: ", vector)

matrix = np.array([[1, 2, 3],
                   [4, 5, 6]])
print("Matix: ", matrix)

tensor = np.array([[[1,2], [3,4]],
                   [[5,6 ], [7, 8]]])
print("Tensor:", tensor)
```

## Array Properties

```python
arr = np.array([[1,2,3],
                [4, 5, 6]])
print("Shape ", arr.shape)
print("Dimension", arr.ndim)
print("Size ", arr.size)
print("DataType", arr.dtype)
```

## Arrange Reshaping

```python
arr = np.arange(12)
print("Original Array", arr)
```

```python
reshaped = arr.reshape((3, 4))
print("\n Reshaped Array ", reshaped)
flattened = reshaped.flatten()
print("\n Falttened array ", flattened)


```

```python
raveled = reshaped.ravel() # ravel (return view , instead of copy)
print("\n Raveled array ", raveled)
```

```python
#Transpose
transpose = reshaped.T
print("\n Transposed Array", transpose)
```

## Creating Array from scratch


```python
zeros = np.zeros((3,4))
print("zeros array : \n", zeros)

ones = np.ones((2,3))
print("ones array : \n", ones)

full = np.full((2, 2), 7)
print("full array: \n", full)

random = np.random.random((2,3))
print("random array: \n", random)

sequence = np.arange(0, 11, 2)
print("sequence array: \n", sequence)
```
