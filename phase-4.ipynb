```python
import matplotlib.image as mpimg

# Step 1: Load the PNG image
logo_img = mpimg.imread("numpy-logo.png")  # make sure this filename matches exactly

# Step 2: Save it as a .npy file
np.save("numpy-logo.npy", logo_img)

# Step 3: Load the .npy file
```

```python
try: 
    logo = np.load("numpy-logo.npy")
   
    #Dispaly
    plt.figure(figsize = (10, 5))
    plt.subplot(121)
    plt.imshow(logo)
    plt.title("Numpy logo")
    plt.grid(False)
    
    dark_logo = 1 - logo
    plt.subplot(122)
    plt.imshow(dark_logo)
    plt.title("Numpy Dark Logo")
    plt.grid(False)

except FileNotFoundError:
    print("numpy logo file not found")
```

```python
import numpy as np
import matplotlib.pyplot as plt

try:
    logo = np.load("numpy-logo.npy")

    plt.figure(figsize=(10, 5))

    # Original
    plt.subplot(121)
    plt.imshow(logo)
    plt.title("Numpy Logo")
    plt.axis("off")

    # Dark version
    if logo.dtype == np.uint8:  # values 0–255
        dark_logo = 255 - logo
    else:  # values 0–1
        dark_logo = 1 - logo

    plt.subplot(122)
    plt.imshow(dark_logo)
    plt.title("Numpy Dark Logo")
    plt.axis("off")

    plt.show()

except FileNotFoundError:
    print("numpy logo file not found")
```
