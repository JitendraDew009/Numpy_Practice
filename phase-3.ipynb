## Advance operation with Business examples

```python
# Data structure [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000], # Paradise Biryani
    [2, 120000, 140000, 160000, 190000], # Beijing Bites
    [3, 20000, 230000, 260000, 300000], # Pizza Hub
    [4, 180000, 210000, 240000, 270000], # Burger Point
    [5, 160000, 185000, 205000, 230000] # Chai Point
])
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
print("====Zomato Sales Annalyis===")
print("\n Sales data shape", sales_data.shape)
print("\n Sample data for 1st 3 resturant", sales_data[0:3])
```

```python
print("\n Sample data for all resturants", sales_data[:,1:]) #[all_row:all_col, 1(start from row 1):all_col]
```

```python
# sum of array in 1 row 
print(np.sum(sales_data, axis = 0))
```

```python
#excluding restaurant coloumn only
# total sales per year
year_total = np.sum(sales_data[:, 1:], axis = 0) #axis = 0 for coloumn
print("total sales per year :",year_total)
```

```python
#Minimum sales per restaurant
min_sales = np.min(sales_data[:, 1:], axis =1) #row axis = 1
print("Minimum sales per restaurant :",min_sales)
```

```python
# Maximum Sales per year
max_sales = np.max(sales_data[:, 1:], axis = 0) 
print("Maximum Sales per year: ",max_sales)
```

```python
#Average sales per year
avg_sales = np.mean(sales_data[:,1:], axis =0)
print("Average sales per year :",avg_sales)
```

```python
# cumulative sum
cum_sum = np.cumsum(sales_data[:, 1:], axis = 1)
print("cumulative sum sales",cum_sum)
```

```python
plt.figure(figsize =(10,7))
plt.plot(np.mean(cum_sum, axis = 0))
plt.title("Average cumulative sales accors all restaurant ")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1, 2, 3, 4, 5])
vector2 = np.array([6, 7, 8, 9, 10])
print("Vector addition: ", vector1 + vector2)
print("Vector subtraction: ", vector1 - vector2)
print("Vector Multiplication: ", vector1 * vector2)
print("\n Dot Product", np.dot(vector1, vector2))
angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1)* np.linalg.norm(vector2))) 
print("angle :",angle)
```
