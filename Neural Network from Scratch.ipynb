```python
# Activation Function (Sigmoid) : Output between 0 and 1
import numpy as np

def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
# Dataset
X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])

y = np.array([[0], [1], [1], [0]])   # XOR problem
```

```python
# Initialize Weights
np.random.seed(42)

W1 = np.random.rand(2, 2)   # input → hidden
b1 = np.zeros((1, 2))

W2 = np.random.rand(2, 1)   # hidden → output
b2 = np.zeros((1, 1))
```

```python
# Forward Propagation
def forward(X):
    Z1 = np.dot(X, W1) + b1
    A1 = sigmoid(Z1)

    Z2 = np.dot(A1, W2) + b2
    A2 = sigmoid(Z2)

    return A1, A2
```

```python
# Loss Function
def compute_loss(y, y_pred):
    return np.mean((y - y_pred) ** 2)
```

```python
# Backpropagation
def backward(X, y, A1, A2):
    m = X.shape[0]

    dZ2 = A2 - y
    dW2 = (1/m) * np.dot(A1.T, dZ2)
    db2 = (1/m) * np.sum(dZ2, axis=0)

    dZ1 = np.dot(dZ2, W2.T) * A1 * (1 - A1)
    dW1 = (1/m) * np.dot(X.T, dZ1)
    db1 = (1/m) * np.sum(dZ1, axis=0)

    return dW1, db1, dW2, db2
```

```python
# Training Loop
learning_rate = 0.1
epochs = 10000

for i in range(epochs):
    A1, A2 = forward(X)

    loss = compute_loss(y, A2)

    dW1, db1, dW2, db2 = backward(X, y, A1, A2)

    W1 -= learning_rate * dW1
    b1 -= learning_rate * db1

    W2 -= learning_rate * dW2
    b2 -= learning_rate * db2

    if i % 1000 == 0:
        print(f"Epoch {i}, Loss: {loss}")
```

```python
# Final Prediction
_, predictions = forward(X)
print(predictions)
```
