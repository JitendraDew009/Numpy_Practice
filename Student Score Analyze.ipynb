# Student Score Analyze

```python
import numpy as np

# Rows = students, Columns = subjects
data = np.array([
    [85, 90, 78],
    [60, 70, 65],
    [95, 88, 92],
    [40, 50, 45],
    [70, 75, 80]
])

print(data)
```

### Goal

Build a small system using NumPy that:

- Analyzes student scores
- Cleans data
- Computes statistics
- Applies filtering


```python
# Average marks per student
student_avg1 = np.mean(data, axis=1)

# Average per subject
subject_avg2 = np.mean(data, axis=0) 

print(student_avg1) # Column wise
print(subject_avg2) # Row wise
```

```python
#Who is topper ?
toppers = data[student_avg > 80]
print(toppers)
```
